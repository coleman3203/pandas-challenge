```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
heroes_df = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#TOTAL PLAYERS
player_count = len(heroes_df["SN"].unique())
TOTAL_PLAYERS = pd.DataFrame({"Total Players":[player_count]})
TOTAL_PLAYERS = TOTAL_PLAYERS.style.set_properties(**{'text-align': 'left'})
TOTAL_PLAYERS.set_table_styles([dict(selector='th', props=[('text-align', 'left')])])
TOTAL_PLAYERS
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#PURCHASING ANALYSIS (TOTAL)
#Unique Items
unique_items = len(heroes_df["Item ID"].unique())
#Average Purchase Price
avg_pp = heroes_df["Price"].mean()
#Total Unique Purchases
unique_purchases = len(heroes_df["Purchase ID"].unique())
#Total Revenue
total_revenue = heroes_df["Price"].sum()

PURCHASING_ANALYSIS = pd.DataFrame({"Number of Unique Items": [unique_items],
                             "Average Price": [avg_pp],
                             "Number of Purchases": [unique_purchases],
                             "Total Revenue":[total_revenue]})
PURCHASING_ANALYSIS["Average Price"] = PURCHASING_ANALYSIS["Average Price"].map("${:.2f}".format)
PURCHASING_ANALYSIS["Total Revenue"] = PURCHASING_ANALYSIS["Total Revenue"].map("${:.2f}".format)

PURCHASING_ANALYSIS
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#GENDER DEMOGRAPHICS
#Count of Male Players
male_df = heroes_df.loc[heroes_df["Gender"] == "Male"]
male_count = len(male_df["SN"].unique())
#Count of Female Players
female_df = heroes_df.loc[heroes_df["Gender"] == "Female"]
female_count = len(female_df["SN"].unique())
#Count of Other/Non-Disclosed
other_df = heroes_df.loc[heroes_df["Gender"] == "Other / Non-Disclosed"]
other_count = len(other_df["SN"].unique())
#Percentage of Male Players
male_perc = male_count/player_count*100
#Percentage of Female Players
female_perc = female_count/player_count*100
#Percentage of Other / Non-Disclosed
other_perc = other_count/player_count*100

GENDER_DEMOGRAPHICS = pd.DataFrame({"":["Male", "Female", "Other / Non Disclosed"],
                                    "Total Count": [male_count, female_count, other_count],
                                    "Percentage of Players": [male_perc, female_perc, other_perc]})
GENDER_DEMOGRAPHICS.set_index("",inplace=True)
GENDER_DEMOGRAPHICS["Percentage of Players"] = GENDER_DEMOGRAPHICS["Percentage of Players"].map("{:.2f}%".format)

GENDER_DEMOGRAPHICS
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#PURCHASING ANALYSIS (GENDER)
gender_df = heroes_df
gender_group = gender_df.groupby(["Gender"])
gender_count = gender_group["Purchase ID"].count()
gender_avgPP = gender_group["Price"].mean()
gender_df.groupby("Gender")["SN"].nunique()
gender_total = gender_group["Price"].sum()
gender_APTpPbAG = gender_group["Price"].sum()/gender_df.groupby("Gender")["SN"].nunique()
GENDER_SUMMARY = pd.DataFrame({"Purchase Count": gender_count,
                             "Average Purchase Price": gender_avgPP,
                             "Total Purchase Value": gender_total,
                             "Average Total Purchase per Person":gender_APTpPbAG})
GENDER_SUMMARY["Average Purchase Price"] = GENDER_SUMMARY["Average Purchase Price"].map("${:.2f}".format)
GENDER_SUMMARY["Total Purchase Value"] = GENDER_SUMMARY["Total Purchase Value"].map("${:.2f}".format)
GENDER_SUMMARY["Average Total Purchase per Person"] = GENDER_SUMMARY["Average Total Purchase per Person"].map("${:.2f}".format)
GENDER_SUMMARY
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_df = heroes_df
bins = [0,9,14,19,24,29,34,39,50]
age_bins = ["<10","10-14","15-19","20-24","25-29","30-34","35-39",">40"]
age_df["Age Ranges"] = pd.cut(age_df["Age"], bins, labels=age_bins)
bin_df = age_df.groupby("Age Ranges")
age_range_count = heroes_df.groupby("Age Ranges")["SN"].nunique()
age_range_perc = age_range_count/player_count*100
AGE_SUMM = pd.DataFrame({"Total Count":age_range_count,
                     "Percentage of Players":age_range_perc})
AGE_SUMM["Percentage of Players"] = AGE_SUMM["Percentage of Players"].map("{:.2f}%".format)
AGE_SUMM = AGE_SUMM.style.set_properties(**{'text-align': 'left'})
AGE_SUMM.set_table_styles([dict(selector='th', props=[('text-align', 'left')])])
AGE_SUMM
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_df = heroes_df
bins = [0,9,14,19,24,29,34,39,50]
age_bins = ["<10","10-14","15-19","20-24","25-29","30-34","35-39",">40"]
age_df["Age Bins"] = pd.cut(age_df["Age"], bins, labels=age_bins)

grouped_age_df = age_df.groupby(["Age Bins"])
agebin_count = age_df["Age Bins"].value_counts()
avgpp = bin_df["Price"].mean()
age_totalvalue = bin_df["Price"].sum()
age_APTpPbAG = bin_df["Price"].sum()/heroes_df.groupby("Age Bins")["SN"].nunique()

AGE_ANALYSIS = pd.DataFrame({"Purchase Count": agebin_count,
                             "Average Purchase Price": avgpp,
                             "Total Purchase Value": age_totalvalue,
                             "Average Total Purchase per Person":age_APTpPbAG})
AGE_ANALYSIS["Average Purchase Price"] = AGE_ANALYSIS["Average Purchase Price"].map("${:.2f}".format)
AGE_ANALYSIS["Total Purchase Value"] = AGE_ANALYSIS["Total Purchase Value"].map("${:.2f}".format)
AGE_ANALYSIS["Average Total Purchase per Person"] = AGE_ANALYSIS["Average Total Purchase per Person"].map("${:.2f}".format)

AGE_ANALYSIS
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
topspenders = heroes_df.groupby(["SN"])
spender_count = topspenders["SN"].count()
spender_average = topspenders["Price"].mean()
spender_sum = topspenders["Price"].sum()
TOP5_SPENDERS = pd.DataFrame({ "Purchase Count": spender_count,
                             "Average Purchase Price": spender_average,
                             "Total Purchase Value": spender_sum})
TOP5_SPENDERS = TOP5_SPENDERS.sort_values("Total Purchase Value", ascending=False)
TOP5_SPENDERS["Total Purchase Value"] = TOP5_SPENDERS["Total Purchase Value"].map("${:.2f}".format)
TOP5_SPENDERS["Average Purchase Price"] = TOP5_SPENDERS["Average Purchase Price"].map("${:.2f}".format)

TOP5_SPENDERS.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_df = heroes_df
item_group_both = item_df.groupby(["Item ID","Item Name"])
item_group_count = item_group_both["Item ID"].count()
item_group_sum = item_group_both["Price"].sum()
item_group_price = item_group_both["Price"].max()
ITEM_SUMMARY = pd.DataFrame({"Purchase Count":item_group_count,
                     "Item Price":item_group_price,
                     "Total Purchase Value":item_group_sum})
POPULAR_SUMMARY = ITEM_SUMMARY.sort_values("Purchase Count", ascending=False)
POPULAR_SUMMARY["Total Purchase Value"] = POPULAR_SUMMARY["Total Purchase Value"].map("${:.2f}".format)
POPULAR_SUMMARY["Item Price"] = POPULAR_SUMMARY["Item Price"].map("${:.2f}".format)

POPULAR_SUMMARY.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
PROFIT_SUMMARY = ITEM_SUMMARY.sort_values("Total Purchase Value", ascending=False)
PROFIT_SUMMARY["Total Purchase Value"] = PROFIT_SUMMARY["Total Purchase Value"].map("${:.2f}".format)
PROFIT_SUMMARY["Item Price"] = PROFIT_SUMMARY["Item Price"].map("${:.2f}".format)
PROFIT_SUMMARY.head()
```
